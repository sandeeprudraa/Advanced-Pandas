```python
df = pd.read_csv("Advanced_pandas_sales_dataset.csv")

display(df.head())
```

Q1. Dataset Exploration

```python
# First 10 Rows
display(df.head(10))

# Dataset Structure
df.info()

# Number of Rows and Columns
print("Shape of Dataset:", df.shape)

# Data Types
print("\nData Types")
print(df.dtypes)
```

Q2. Sales Filtering Analysis

```python
print("Sales Greater Than 3500")

display(df[df["Sales"] > 3500])

print("Region = East and Sales > 2500")

display(df[(df["Region"]=="East") & (df["Sales"]>2500)])
```

Q3. Customer Purchase Analysis

```python
top_customers = df.groupby("Customer")["Sales"].sum().sort_values(ascending=False)

display(top_customers.head(5))
```

Q4. Regional Sales Performance

```python
region_sales = df.groupby("Region")["Sales"].sum()

display(region_sales)

highest_region = region_sales.idxmax()

print("Highest Revenue Region:", highest_region)
```

Q5. Product Performance Analysis

```python
product_quantity = df.groupby("Product")["Quantity"].sum()

display(product_quantity)

print("Most Popular Product:")

print(product_quantity.idxmax())
```

Q6. Multi-Level Grouping

```python
region_category = df.groupby(["Region","Category"])["Sales"].sum()

display(region_category)
```

Q7. Pivot Table Creation

```python
pivot = pd.pivot_table(
    df,
    values="Sales",
    index="Region",
    columns="Product",
    aggfunc="sum",
    fill_value=0
)

display(pivot)

print("Region Selling Maximum Laptops:")

print(pivot["Laptop"].idxmax())
```

Q8. Profitability Analysis

```python
category_profit = df.groupby("Category")["Profit"].mean()

display(category_profit)

print("Most Profitable Category:")

print(category_profit.idxmax())
```

Q9. Discount Impact Study

```python
discount_profit = df.groupby("Discount")["Profit"].mean()

display(discount_profit)
```

Q10. End-to-End Data Wrangling

```python
filtered = df[df["Sales"] > 3000]

grouped = filtered.groupby(["Region","Product"])["Sales"].sum().reset_index()

pivot_table = grouped.pivot(
    index="Region",
    columns="Product",
    values="Sales"
)

display(pivot_table.fillna(0))
```
